```python
import os
import glob
from pprint import pprint
from datetime import datetime

import numpy as np
import pandas as pd
import graphviz
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import (
    LinearRegression,
    LogisticRegression,
    LogisticRegressionCV,
    Ridge,
    Lasso,
)
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.pipeline import Pipeline
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import KFold, train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier, export_graphviz
from sklearn import metrics
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import KFold, train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier, export_graphviz
from sklearn.metrics import (
    confusion_matrix,
    f1_score,
    accuracy_score,
    classification_report,
    recall_score,
    precision_score,
    roc_auc_score,
    roc_curve,
    auc
)
from sklearn.ensemble import GradientBoostingClassifier, GradientBoostingRegressor
from sklearn.neural_network import MLPClassifier, MLPRegressor
from sklearn.svm import SVC, SVR
from xgboost import XGBClassifier, XGBRegressor
from sklearn.tree import DecisionTreeClassifier, plot_tree
import matplotlib.pyplot as plt

from tensorflow.keras.layers import Dense
from tensorflow.keras.models import Sequential
from tensorflow.keras.metrics import AUC
```

```python
features = ['YearsWithCurrManager', 'Education', 'NumCompaniesWorked', 'EnvironmentSatisfaction', 'JobSatisfaction']
X = df[features]
y = df['Group']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)

# Tiền xử lý dữ liệu
preprocessor = ColumnTransformer(
    transformers=[
        ('num', StandardScaler(), features)
    ])
```

```python
le = LabelEncoder()
y_train = le.fit_transform(y_train)
y_test = le.transform(y_test)

# Khởi tạo DataFrame để lưu kết quả
results_df = pd.DataFrame(columns=['Model', 'Accuracy', 'Precision', 'Recall', 'F1 Score', 'AUC'])

models = {
    'Logistic Regression': LogisticRegression(random_state=42),
    'Decision Tree': DecisionTreeClassifier(random_state=42),
    'Random Forest': RandomForestClassifier(random_state=42),
    'Gradient Boosting': GradientBoostingClassifier(random_state=42),
    'MLP Neural Network': MLPClassifier(random_state=42, hidden_layer_sizes=(100,), max_iter=1000),
    'XGBoost': XGBClassifier(random_state=42)
}

# Huấn luyện và đánh giá từng mô hình
for model_name, model in models.items():
    pipeline = Pipeline(steps=[('preprocessor', preprocessor), ('model', model)])
    pipeline.fit(X_train, y_train)  # Đảm bảo rằng y_train đã được chuyển đổi

    # Dự đoán trên tập kiểm tra
    y_pred = pipeline.predict(X_test)
    y_pred_proba = pipeline.predict_proba(X_test)[:, 1]  # Chỉ tính xác suất cho lớp '1'

    # Tính toán các chỉ số đánh giá
    accuracy = accuracy_score(y_test, y_pred)
    precision = precision_score(y_test, y_pred, pos_label=1)  # Đảm bảo rằng pos_label là số nguyên
    recall = recall_score(y_test, y_pred, pos_label=1)
    f1 = f1_score(y_test, y_pred, pos_label=1)
    roc_auc = roc_auc_score(y_test, y_pred_proba)

    # Thêm kết quả vào DataFrame
    results_df.loc[len(results_df)] = {
        'Model': model_name,
        'Accuracy': accuracy,
        'Precision': precision,
        'Recall': recall,
        'F1 Score': f1,
        'AUC': roc_auc
    }
# Hiển thị kết quả
print(results_df)
```

##Recursive Feature Elimination

```python
from yellowbrick.features import RFECV
from sklearn.model_selection import StratifiedKFold
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
import xgboost as xgb

# Chắc chắn rằng y_train và y_test là kiểu số nguyên
y_train = y_train.astype(int)
y_test = y_test.astype(int)

# Cài đặt cross-validation
cv = StratifiedKFold(5)

# Các biến lưu tham số tốt nhất từ Grid Search
best_params_lr = lr_grid.best_params_
best_params_dt = dt_grid.best_params_
best_params_rf = rf_grid.best_params_
best_params_gb = gb_grid.best_params_
best_params_xgb = xgb_grid.best_params_

# Danh sách các mô hình và tham số tương ứng (không bao gồm MLPClassifier)
models_and_params = [
    (LogisticRegression, best_params_lr),
    (DecisionTreeClassifier, best_params_dt),
    (RandomForestClassifier, best_params_rf),
    (GradientBoostingClassifier, best_params_gb),
    (xgb.XGBClassifier, best_params_xgb)  # Đảm bảo xgb.XGBClassifier được gọi đúng
]

# Duyệt qua từng mô hình và thực hiện RFECV
for model_cls, params in models_and_params:
    model = model_cls(**params)
    oz = RFECV(estimator=model, cv=cv, scoring='accuracy')
    oz.fit(X=X_train, y=y_train)
    print(f"Model: {model_cls.__name__}")
    print("Features selected:", oz.support_)
    oz.poof()  # Hiển thị biểu đồ
```
